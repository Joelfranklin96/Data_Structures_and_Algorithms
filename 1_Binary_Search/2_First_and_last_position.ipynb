```python
def binary_search(lo,hi,condition):
    
    while lo<=hi:
        
        mid = (lo + hi)//2
        
        if condition(mid) == 'found':
            return mid
        elif condition(mid) == 'left':
            hi = mid - 1
        else:
            lo = mid + 1 
    
    return -1
    
def first_position(cards,query):
    
    def condition(mid):
        
        mid_number = cards[mid]
        
        if query == mid_number:
            if mid>0 and cards[mid-1] == query:
                return 'left'
            else:
                return 'found'
        elif query > mid_number:
            return 'right'
        else:
            return 'left'
    
    return binary_search(0,len(cards)-1,condition)

def last_position(cards,query):
    
    def condition(mid):
        
        mid_number = cards[mid]
        
        if query == mid_number:
            if mid < len(cards)-1 and cards[mid+1] == query:
                return 'right'
            else:
                return 'found'
        elif query > mid_number:
            return 'right'
        else:
            return 'left'
    
    return binary_search(0,len(cards)-1,condition)

def first_and_last_position(cards,query):
    return first_position(cards,query),last_position(cards,query)
```

```python
cards = [2,3,4,4,4,4,4,4,5,7,9,10,12]
query = 4
first_and_last_position(cards,query)
```
