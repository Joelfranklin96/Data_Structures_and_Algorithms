```python
def helper(cards,query,mid):
    
    mid_number = cards[mid]
    
    if query == mid_number:
        if mid>0 and cards[mid-1] == query:
            return 'left'
        else:
            return 'found'
    elif query > mid_number:
        return 'right'
    else:
        return 'left'

def binary_search(cards,query):
    
    lo = 0
    hi = len(cards) - 1
    
    while lo <= hi:
        
        mid = (hi + lo)//2
        
        if helper(cards,query,mid) == 'found':
            return mid
        elif helper(cards,query,mid) == 'left':
            hi = mid - 1
        else:
            lo = mid + 1
    
    return -1
```
