Bubble Sort and Insertion Sort algorithms.

```python
# Bubble Sort

def bubble_sort(num):
    
    num = list(num)
    
    for i in range(len(num)-1):
        
        for j in range(len(num)-1):
            
            if num[j] > num[j+1]:
                num[j],num[j+1] = num[j+1],num[j]
    
    return num
```

```python
bubble_sort([4,2,5,3,6,6,5,5,8,8,2,2,4])
```

```python
# Insertion Sort

def insertion_sort(num):
    
    for i in range(len(num)):
        
        cur = num.pop(i)
        
        j = i - 1
        
        while j>=0 and num[j] > cur:
            j = j - 1
        
        num.insert(j+1,cur)
    
    return num
```

```python
insertion_sort([3,3,2,4,5,3,2,1,6])
```
