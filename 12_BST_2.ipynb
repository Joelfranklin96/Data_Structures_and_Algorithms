```python
class BinaryTreeNode():
    
    def __init__(self,data):
        
        self.data = data
        self.left = None
        self.right = None
```

```python
def binary_tree_input():
    
    root_data = int(input())
    
    if root_data == -1:
        return None
    
    root = BinaryTreeNode(root_data)
    root.left = binary_tree_input()
    root.right = binary_tree_input()
    
    return root
```

```python
import queue

def levelwise_input():
    
    root = None
    print("Please enter the root data")
    key = int(input())
    
    q = queue.Queue()
    
    if key == -1:
        return None
    else:
        root = BinaryTreeNode(key)
        q.put(root)
        
        while not(q.empty()):
            current_node = q.get()
            
            print("Please enter the left child of {}".format(current_node.data))
            key = int(input())
            
            if key != -1:
                left = BinaryTreeNode(key)
                q.put(left)
                current_node.left = left
            else:
                current_node.left = None
            
            print("Please enter the right child of {}".format(current_node.data))
            key = int(input())
            
            if key != -1:
                right = BinaryTreeNode(key)
                q.put(right)
                current_node.right = right
            else:
                current_node.right = None
        
        return root    
```

```python
# Create a BST class which has 4 methods namely 'Insert', 'Delete' and 'Search', 'Count' and 'Print'

# Delete
# Given an element, remove that element from the BST. 
# If the element which is to be deleted has both children, replace that with the minimum element from right sub-tree and then remove the minimum element node
```

```python
class BST:
    def __init__(self):
        self.root = None
        self.numNodes = 0
    
    def insert_iterative(self, data):
        if self.root is None:
            self.root = BinaryTreeNode(data)
        else:
            node = self.root
            while True:
                if data < node.data:
                    if node.left is None:
                        node.left = BinaryTreeNode(data)
                        break
                    node = node.left
                else:
                    if node.right is None:
                        node.right = BinaryTreeNode(data)
                        break
                    node = node.right
    
    def insert_recursive(self, data):
        if self.root is None:
            self.root = BinaryTreeNode(data)
        else:
            self.insert_recursive_helper(self.root, data)
    
    def insert_recursive_helper(self, root, data):
        if root is None:
            root = BinaryTreeNode(data)
        else:
            if data < root.data:
                if root.left is None:
                    root.left = BinaryTreeNode(data)
                else:
                    self.insert_recursive_helper(root.left, data)
            else:
                if root.right is None:
                    root.right = BinaryTreeNode(data)
                else:
                    self.insert_recursive_helper(root.right, data)
    
    def insert_recursive_helper_2(self, root, data):
        if root is None:
            return BinaryTreeNode(data)
        else:
            if data < root.data:
                root.left = self.insert_recursive_helper_2(root.left, data)
                return root
            else:
                root.right = self.insert_recursive_helper_2(root.right, data)
                return root
    
    def delete_recursive(self, data):
        self.delete_recursive_helper_2(self.root, data)
    
    def delete_recursive_helper(self, root, data):
        
        if root is None:
            return
        
        if root.left is None and root.right is None:
            return
        
        if root.left is not None and root.left.data == data:
            if root.left.left is None and root.left.right is None:
                root.left = None
            elif root.left.left is None:
                root.left = root.left.right
            elif root.left.right is None:
                root.left = root.left.left
            else:
                temp = root.left
                node = BinaryTreeNode(self.minimum_helper(root.left.right))
                root.left = node
                node.left = temp.left
                node.right = temp.right
        elif root.right is not None and root.right.data == data:
            if root.right.left is None and root.right.right is None:
                root.right = None
            elif root.right.left is None:
                root.right = root.right.right
            elif root.right.right is None:
                root.right = root.right.left
            else:
                temp = root.right
                node = BinaryTreeNode(self.minimum_helper(root.right.right))
                root.right = node
                node.left = temp.left
                node.right = temp.right
        else:
            if data <= root.data:
                self.delete_recursive_helper(root.left, data)
            else:
                self.delete_recursive_helper(root.right, data)
    
    def delete_recursive_helper_2(self, root, data):
        if root is None:
            return None
        if data < root.data:
            root.left = self.delete_recursive_helper_2(root.left, data)
            return root
        elif data > root.data:
            root.right = self.delete_recursive_helper_2(root.right, data)
            return root
        else:
            if root.left is None and root.right is None:
                return None
            elif root.left is None:
                return root.right
            elif root.right is None:
                return root.left
            else:
                replacement = self.minimum_helper(root.right)
                root.data = replacement
                root.right = self.delete_recursive_helper_2(root.right, replacement)
                return root
                           
    def minimum_helper(self, root):
        
        if root.left is None:
            return root.data
        else:
            return minimum_helper(root.left)        
    
    def search_iterative(self, data):
        node = self.root
        while node is not None:
            if data == node.data:
                return True
            elif data < node.data:
                node = node.left
            else:
                node = node.right
        return False
    
    def search_recursive(self, data):
        return self.search_recursive_helper(self.root, data)
    
    def search_recursive_helper(self, node, data):
        if node is None:
            return False
        if node.data == data:
            return True
        elif data < node.data:
            return self.search_recursive_helper(node.left, data)
        else:
            return self.search_recursive_helper(node.right, data)
    
    def print_tree(self):
        self.print_tree_helper(self.root)
    
    def print_tree_helper(self, node):
        if node is None:
            return
        print("{}:".format(node.data), end = "")
        if node.left is not None and node.right is not None:
            print("L:{},R:{}".format(node.left.data, node.right.data))
        elif node.left is not None:
            print("L:{},R:-1".format(node.left.data))
        elif node.right is not None:
            print("L:-1,R:{}".format(node.right.data))
        else:
            print("L:-1,R:-1")
        
        self.print_tree_helper(node.left)
        self.print_tree_helper(node.right)
```

```python
# Time complexity

#1) Search
#Time complexity = O(h) where h is the height of the binary search tree

#2) Insert
#Time complexity = O(h) where h is the height of the binary search tree

#3) Delete
# Time complexity = O(h) where h is the height of the binary search tree

# Step 1 - Work done in finding the node with the given element = p (We traverse p nodes)
# Step 2 - Work done in finding the minimum of the right subtree = h - p (We traverse h - p nodes to find the minimum)
# Step 3 - Work done in deleting the minimum from the right subtree can not be more than (h - p).
# Because the node with the minimum value will not have a left. It will either have 0 children (work done = h - p)
# or it have 1 child (work done is less than h - p)

# Overall time complexity of delete operation in a BST = O(h)
```
