Alice has some cards with numbers written on them. She arranges the cards in increasing order and lays them out face down in a sequence on a table. She challenges Bob to pick out the card containing a given number by turning over as few cards as possible.

The inputs are a deck of cards in increasing order and a query number to be searched. The output is the lowest index of the query number.

```python
def helper(cards,query,mid):
    
    mid_number = cards[mid]
    
    if query == mid_number:
        if mid>0 and cards[mid-1] == query:
            return 'left'
        else:
            return 'found'
    elif query > mid_number:
        return 'right'
    else:
        return 'left'

def binary_search(cards,query):
    
    lo = 0
    hi = len(cards) - 1
    
    while lo <= hi:
        
        mid = (hi + lo)//2
        
        if helper(cards,query,mid) == 'found':
            return mid
        elif helper(cards,query,mid) == 'left':
            hi = mid - 1
        else:
            lo = mid + 1
    
    return -1
```

```python
cards = [2,4,4,4,4,4,5,8,10,13]
query = 4
binary_search(cards,query)
```
