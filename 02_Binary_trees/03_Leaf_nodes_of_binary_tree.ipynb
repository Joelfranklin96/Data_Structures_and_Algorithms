Write a function to return the key values of leaf nodes of a binary tree.

```python
class treenode:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
```

```python
def tuple_to_tree(data):
    if isinstance(data,tuple):
        node = treenode(data[1])
        node.left = tuple_to_tree(data[0])
        node.right = tuple_to_tree(data[2])
    elif data is not None:
        node = treenode(data)
    else:
        node = None
    
    return node
```

```python
node = tuple_to_tree((((None,1,6),3,(3,4,2)),2,((4,3,None),5,(6,7,8))))
```

```python
def leaves_binary_tree(node):
    if node is None:
        return []
    elif node.left is None and node.right is None:
        return [node.key]
    return leaves_binary_tree(node.left) + leaves_binary_tree(node.right)
```

```python
leaves_binary_tree(node)
```
