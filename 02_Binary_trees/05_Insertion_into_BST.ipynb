Write a function to insert a new node into a Binary Search Tree.

```python
class TreeNode:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
```

```python
def tuple_to_tree(data):
    if isinstance(data,tuple):
        node = TreeNode(data[1])
        node.left = tuple_to_tree(data[0])
        node.right = tuple_to_tree(data[2])
    elif data is not None:
        node = TreeNode(data)
    else:
        node = None
    
    return node
```

```python
def tree_to_tuple(node):
    x = [0,0,0]
    if isinstance(node,TreeNode) and (node.left is not None or node.right is not None):
        x[0] = tree_to_tuple(node.left)
        x[1] = node.key
        x[2] = tree_to_tuple(node.right)
    elif node is None:
        return None
    else:
        return node.key
    
    return tuple(x)
```

```python
node = tuple_to_tree(((2,3,4),7,(9,11,12)))
```

```python
def insert(node,key):
    
    if node is None:
        node = TreeNode(key)
    elif key < node.key:
        node.left = insert(node.left,key)
    else:
        node.right = insert(node.right,key)
    return node
```

```python
new_node = insert(node,5)
```

```python
tree_to_tuple(new_node)
```
