Write a function to retrieve all the key-values pairs stored in a BST in the sorted order of keys.

The nodes can be listed in sorted order by performing an inorder traversal of the BST.

```python
class BSTNode:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.left = None
        self.right = None
```

```python
node = BSTNode(7,'seven')
node.left = BSTNode(3,'three')
node.left.left = BSTNode(2,'two')
node.left.right = BSTNode(4,'four')
node.right = BSTNode(11,'eleven')
node.right.left = BSTNode(9,'nine')
node.right.right = BSTNode(12,'twelve')
```

```python
def traversal(node):
    if node is None:
        return []
    return traversal(node.left) + [(node.key,node.value)] + traversal(node.right)
```
