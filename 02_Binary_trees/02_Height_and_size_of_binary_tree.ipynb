Write a function to calculate the height/depth and size of binary tree. The height/depth of a binary tree is defined as the length of the longest path from its root node to a leaf. The size of a binary tree is defined as the number of nodes in a binary tree.

```python
class treenode:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
```

```python
def tuple_to_tree(data):
    if isinstance(data,tuple):
        node = treenode(data[1])
        node.left = tuple_to_tree(data[0])
        node.right = tuple_to_tree(data[2])
    elif data is not None:
        node = treenode(data)
    else:
        node = None
    
    return node
```

```python
node = tuple_to_tree((((None,1,6),3,(3,4,2)),2,((None,3,4),5,(6,7,8))))
```

```python
def tree_height(node):
    if node is None:
        return 0
    return 1 + max(tree_height(node.left), tree_height(node.right))
```

```python
tree_height(node)
```

```python
def tree_size(node):
    if node is None:
        return 0
    return 1 + tree_size(node.left) + tree_size(node.right)
```

```python
tree_size(node)
```
